```python
import pandas as pd
```

```python
all_dfs = []

for i in range(1, 62):
    df = pd.read_csv(f"../datasets/raw_dataset/urls_{i}.csv")
    all_dfs.append(df)

combined_df = pd.concat(all_dfs, ignore_index=True)
print("Combined DataFrame shape:", combined_df.shape)
```

```python
print("Malicious URLs:", combined_df[combined_df["label"] == 1].shape[0])
print("Non-Malicious URLs:", combined_df[combined_df["label"] == 0].shape[0])
```

```python
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
combined_df.head(50)
```

```python
chunk_size = 100_000
output_dir = "../datasets/training_dataset_with_features/"

for i in range(0, len(combined_df), chunk_size):
    chunk = combined_df.iloc[i:i + chunk_size]
    chunk.to_csv(f"{output_dir}urls_{i // chunk_size + 1}.csv", index=False)
    print(f"Saved: urls_{i // chunk_size + 1}.csv with {len(chunk)} rows")
```

```python
all_dfs = []

for i in range(1, 62):
    df = pd.read_csv(f"../datasets/training_dataset_with_features/urls_{i}.csv")
    all_dfs.append(df)

combined_df = pd.concat(all_dfs, ignore_index=True)
print("Combined DataFrame shape:", combined_df.shape)
```

```python
combined_df
```

```python
train_rows = 4800000
val_rows = 600000
test_rows = 693263  

train_df = combined_df.iloc[:train_rows]
val_df = combined_df.iloc[train_rows:train_rows + val_rows]
test_df = combined_df.iloc[train_rows + val_rows:train_rows + val_rows + test_rows]

print("Train shape:", train_df.shape)
print("Validation shape:", val_df.shape)
print("Test shape:", test_df.shape)


train_df.to_csv("../datasets/training_dataset_with_features/train.csv", index=False)
val_df.to_csv("../datasets/training_dataset_with_features/val.csv", index=False)
test_df.to_csv("../datasets/training_dataset_with_features/test.csv", index=False)
```
